# Zero Parameter Structure

## Minimal End-to-End Notebook

This notebook reproduces the main structural calculations from the common definitions

$$
P_{\min}=1+1,\qquad
P_{\max}=3+4,\qquad
P_{\mathrm{mid}}=P_{\max}-P_{\min},
$$

$$
R=2\left(1+\frac{P_{\min}}{24}\right)=\frac{13}{6},
\qquad
S=1+\frac{P_{\max}}{24}=\frac{31}{24}.
$$

It covers cosmology, the electron sector, gravity, quark masses, neutrino masses, and Higgs/electroweak/strong-coupling scale relations.

**No free parameters. No tuning. Only structure.**

## 1. Imports and common structural definitions

The electron mass-energy scale is retained explicitly in eV, MeV, and GeV:

$$
m_e^{(\mathrm{MeV})}=10^{-6}m_e^{(\mathrm{eV})},
\qquad
m_e^{(\mathrm{GeV})}=10^{-9}m_e^{(\mathrm{eV})}.
$$

The structural-transfer notation is

$$
\alpha^2_{[i\to j]}=\alpha^{2(i-j)}.
$$

````python
from fractions import Fraction
import math

c = 299792458       # exact, m/s
h = 6.62607015e-34  # exact, J s
hbar = h / (2 * math.pi)

P_MIN = 1 + 1
P_MAX = 3 + 4
P_MID = P_MAX - P_MIN

R = (Fraction(1) + Fraction(P_MIN, 24)) * 2
S =  Fraction(1) + Fraction(P_MAX, 24)


def alpha2_transfer(alpha_value, i, j):
    return alpha_value ** (2 * (i - j))


def print_table(rows, name_width=20, value_width=22):
    print(f"{'Quantity':<{name_width}} {'Value':>{value_width}} {'Unit':<8}")
    print("-" * (name_width + value_width + 10))
    for name, value, unit, number_format in rows:
        value_text = format(float(value), number_format)
        print(f"{name:<{name_width}} {value_text:>{value_width}} {unit:<8}")


print(f"P_MIN = {P_MIN}")
print(f"P_MAX = {P_MAX}")
print(f"P_MID = {P_MID}")
print(f"R     = {R} = {float(R):.12f}")
print(f"S     = {S} = {float(S):.12f}")
````

## 2. Cosmological density relations

$$
\Omega_b=\frac{1}{3R^2+3R},
\qquad
\Omega_m=3R\Omega_b,
$$

$$
\Omega_\Lambda=3R^2\Omega_b,
\qquad
\Omega_{\mathrm{dm}}=\Omega_m-\Omega_b.
$$

These relations satisfy

$$
\Omega_m^2=3\Omega_\Lambda\Omega_b.
$$

````python
Omega_b  = 1 / (3 * R**2 + 3 * R)
Omega_m  = 3 * R    * Omega_b
Omega_L  = 3 * R**2 * Omega_b
Omega_dm = Omega_m - Omega_b

print_table([
    ("Omega_L", Omega_L, "", ".12f"),
    ("Omega_m", Omega_m, "", ".12f"),
    ("Omega_dm", Omega_dm, "", ".12f"),
    ("Omega_b", Omega_b, "", ".12f"),
])

print()
print("Exact fractions")
print(f"Omega_L  = {Omega_L}")
print(f"Omega_m  = {Omega_m}")
print(f"Omega_dm = {Omega_dm}")
print(f"Omega_b  = {Omega_b}")
print()
print(f"Consistency residual = {float(Omega_m**2 - 3 * Omega_L * Omega_b):.3e}")
````

## 3. Electromagnetic coupling and mass hierarchy

The fixed backbone indices are constructed from

$$
6R=13,\qquad 24S=31,
$$

and corrected through fixed residual terms:

$$
\Psi_i^{\mathrm{eff}}=\Psi_{i0}\pm\delta_i.
$$

The inverse electromagnetic coupling is

$$
\alpha^{-1}
=
4\pi\frac{3R^2}{S}
\left(1+\frac{1}{\Psi_e^{\mathrm{eff}}}\right).
$$

The electron mass-energy path is

$$
\Psi_{m_e}^{*}
=
24\left[\frac{3}{2}RS(6\cdot24)-1\right],
\qquad
\Psi_{m_e}
=
12\Psi_{m_e}^{*}
+
\frac{\Psi_{e0}}{3}.
$$

````python
core_13 = R * 6
core_31 = S * 24

# Structural backbone values Psi_0.
psi_e0   = Fraction(3, 2) * (core_13    * core_31**2 - P_MAX**2) / 3
psi_p0   = Fraction(3, 2) * (core_13**2 * core_31    + P_MAX**2) / 12
psi_n0   = Fraction(3, 2) * (core_13    * core_31    + P_MIN**2) * 12
psi_mu0  = Fraction(psi_p0, P_MIN**2)
psi_tau0 = psi_p0

# Residual structural fluctuations delta_psi.
delta_e   = Fraction(1, 2) + Fraction(1, P_MID**2)
delta_p   = Fraction(6**2) * Fraction(1, psi_e0 + 24 - Fraction(1, 2))
delta_n   = Fraction(2, 3) * Fraction(12*P_MAX - 1, 12*P_MAX + 3)
delta_mu  = Fraction(1, 2) - Fraction(1, P_MAX**2) + Fraction(1, (psi_mu0 * P_MAX**2))
delta_tau = Fraction(1, 1)

psi_e     = psi_e0   - delta_e
psi_p     = psi_p0   + delta_p
psi_n     = psi_n0   - delta_n
psi_mu    = psi_mu0  - delta_mu
psi_tau   = psi_tau0 + delta_tau

alpha_inv =         4*math.pi * (3*R**2 / S) * (1 + 1/psi_e)
mmu_me    = (3/2) * 4*math.pi * (3*R**2 / S) * (1 + 1/psi_mu)
mtau_mmu  = (3/4)             * (8*R    * S) * (1 + 1/psi_tau)
mp_me     =         alpha_inv * (8*R    / S) * (1 - 1/psi_p)
mn_me     =         alpha_inv * (8*R    / S) * (1 - 1/psi_n)
mtau_me   = mtau_mmu * mmu_me

alpha = alpha_inv**-1

# Electron rest energy in MeV.
# The expression is kept close to the structural form used in the paper code.
psi_me_star = 24 * (Fraction(3, 2) * (R * S) * (6 * 24) - 1)
psi_me      = 12 * psi_me_star + Fraction(psi_e0, 3)
me_ev       = Fraction(c, 10**3) ** 2 / (psi_me * (1 + psi_me_star**-2))
me_mev      = me_ev * 10**-6  # MeV
me_gev      = me_ev * 10**-9  # GeV
````

````python
print("Effective structural indices")
print_table([
    ("Psi_e", psi_e, "", ".12f"),
    ("Psi_p", psi_p, "", ".12f"),
    ("Psi_n", psi_n, "", ".12f"),
    ("Psi_mu", psi_mu, "", ".12f"),
    ("Psi_tau", psi_tau, "", ".12f"),
])

print()
print("Coupling and mass ratios")
print_table([
    ("alpha^-1", alpha_inv, "", ".12f"),
    ("alpha", alpha, "", ".12e"),
    ("m_p / m_e", mp_me, "", ".12f"),
    ("m_n / m_e", mn_me, "", ".12f"),
    ("m_mu / m_e", mmu_me, "", ".12f"),
    ("m_tau / m_mu", mtau_mmu, "", ".12f"),
    ("m_tau / m_e", mtau_me, "", ".12f"),
])

print()
print("Electron mass-energy scale")
print_table([
    ("Psi_me*", psi_me_star, "", ".12f"),
    ("Psi_me", psi_me, "", ".12f"),
    ("m_e c^2", me_ev, "eV", ".9f"),
    ("m_e c^2", me_mev, "MeV", ".15f"),
    ("m_e c^2", me_gev, "GeV", ".14e"),
])
````

## 4. Gravity

$$
\Psi_G
=
\frac{12(3R^2)(4R)+2R}{4},
$$

$$
\sqrt{G}
=
\frac{\alpha S}{\pi\Psi_G},
\qquad
G
=
\left(\frac{\alpha S}{\pi\Psi_G}\right)^2.
$$

````python
psi_g      = (12 * (3 * R**2) * (4 * R) + 2 * R) / 4
psi_g0     = R * S**2 * (12 * 24)
psi_g_star = 4 * psi_g - 3 * (1 + 1 / psi_g0)

sqrt_G   = (alpha * S) / (math.pi * psi_g)
G        = sqrt_G**2
alpha_Gp = alpha**24 * mp_me**4 / (1 + 1 / psi_g_star)
alpha_Ge = alpha**24 * mp_me**2 / (1 + 1 / psi_g_star)

g_corr = (1 + 1 / psi_g_star)**(-1/2)
M_Pl   = math.sqrt(hbar * c / G)
M_P    = M_Pl * alpha**12 * mp_me**2 * g_corr
M_E    = M_Pl * alpha**12 * mp_me    * g_corr

print_table([
    ("Psi_G", psi_g, "", ".12f"),
    ("Psi_G*", psi_g_star, "", ".12f"),
    ("sqrt(G)", sqrt_G, "", ".12e"),
    ("G", G, "", ".12e"),
    ("alpha_Gp", alpha_Gp, "", ".12e"),
    ("alpha_Ge", alpha_Ge, "", ".12e"),
    ("M_Pl", M_Pl, "kg", ".12e"),
    ("M_P", M_P, "kg", ".12e"),
    ("M_E", M_E, "kg", ".12e"),
])
````

## 5. Quark masses

$$
\frac{m_q}{m_e}
=
K_q B_q A_q^{n_q},
\qquad
B_q = 3R^2S.
$$

$$
A_q =
\begin{cases}
A_u & (q=u,c,t), \\
A_d & (q=d,s,b),
\end{cases}
\qquad
A_d = \frac{8R}{S},
\qquad
A_u = \pi R A_d.
$$

$$
(K_u,K_d,K_c,K_s,K_t,K_b)
=
\left(
\frac{1}{2R},\,
\frac{1}{2},\,
1+\frac{1}{2},\,
1-\frac{1}{4},\,
2+\frac{1}{2R},\,
2+\frac{1}{2}
\right).
$$


$$
n_q =
\begin{cases}
0 & (u,d), \\
1 & (c,s), \\
2 & (t,b).
\end{cases}
$$

The physical masses use $m_e^{(\mathrm{MeV})}$.

````python
K_q = {
    "u": (    1 / (2 * R)),
    "d": (    1 /  2),
    "c": (1 + 1 /  2),
    "s": (1 - 1 /  4),
    "t": (2 + 1 / (2 * R)),
    "b": (2 + 1 /  2),
}

B_q = (3 * R**2) * S
A_d = (8 * R) / S
A_u = math.pi * R * A_d

m_u = me_mev * K_q["u"] * B_q
m_d = me_mev * K_q["d"] * B_q
m_c = me_mev * K_q["c"] * B_q * A_u
m_s = me_mev * K_q["s"] * B_q * A_d
m_t = me_gev * K_q["t"] * B_q * A_u**2
m_b = me_gev * K_q["b"] * B_q * A_d**2

print_table([
    ("B_q", B_q, "", ".12f"),
    ("A_u", A_u, "", ".12f"),
    ("A_d", A_d, "", ".12f"),
    ("m_u", m_u, "MeV", ".12f"),
    ("m_d", m_d, "MeV", ".12f"),
    ("m_c", m_c, "MeV", ".12f"),
    ("m_s", m_s, "MeV", ".12f"),
    ("m_t", m_t, "GeV", ".12f"),
    ("m_b", m_b, "GeV", ".12f"),
])
````

## 6. Neutrino masses

$$
B_\nu
=
12S\alpha^2_{[2\to0]}
=
12S\alpha^4,
$$

$$
\frac{m_{\nu,n}}{m_e}
=
B_\nu(n+\Phi_n),
\qquad n=0,1,2,
$$

$$
\Phi_0=1,
\qquad
\Phi_1=\frac{1}{3R^2},
\qquad
\Phi_2=\frac{1}{R}.
$$

The dimensionless ratios and the absolute masses in eV are kept separate.

````python
B_nu = 12 * S * alpha2_transfer(alpha, 2, 0)

phi_nu = (
    1,
    1 / (3 * R**2),
    1 / R,
)

# Dimensionless mass ratios m_nu,n / m_e for n = 0, 1, 2.
nu_mass_ratios = tuple(
    B_nu * (n + phi_nu[n])
    for n in range(3)
)

# Absolute neutrino masses in eV.
m_nu1_ev, m_nu2_ev, m_nu3_ev = tuple(
    ratio * me_ev
    for ratio in nu_mass_ratios
)

dm21 = m_nu2_ev**2 - m_nu1_ev**2
dm31 = m_nu3_ev**2 - m_nu1_ev**2
dm32 = m_nu3_ev**2 - m_nu2_ev**2
sum_mnu = m_nu1_ev + m_nu2_ev + m_nu3_ev

print("Dimensionless mass ratios")
print_table([
    ("m1 / m_e", nu_mass_ratios[0], "", ".12e"),
    ("m2 / m_e", nu_mass_ratios[1], "", ".12e"),
    ("m3 / m_e", nu_mass_ratios[2], "", ".12e"),
])

print()
print("Absolute masses and splittings")
print_table([
    ("m1 c^2", m_nu1_ev, "eV", ".12f"),
    ("m2 c^2", m_nu2_ev, "eV", ".12f"),
    ("m3 c^2", m_nu3_ev, "eV", ".12f"),
    ("Sum m_nu", sum_mnu, "eV", ".12f"),
    ("Delta m21^2", dm21, "eV^2", ".12e"),
    ("Delta m31^2", dm31, "eV^2", ".12e"),
    ("Delta m32^2", dm32, "eV^2", ".12e"),
])
````

## 7. Higgs, electroweak, and strong-coupling scales

$$
B_H
=
12R\alpha^2_{[1\to2]}
=
12R\alpha^{-2},
$$

$$
\Psi_H
=
12\left[4(3R^2)+3\right]
=
712,
\qquad
\Psi_v
=
3^2\Psi_H
=
6408.
$$

$$
\frac{m_H}{m_e}
=
\frac{1}{2}B_H+\Psi_H,
\qquad
\frac{v}{m_e}
=
B_H-\Psi_v,
$$

$$
y_e=\frac{\sqrt{2}}{v/m_e},
$$

$$
\sin^2\theta_W
=
2\pi\alpha R^2(1+3^{-3}),
\qquad
\cos\theta_W
=
\sqrt{1-\sin^2\theta_W},
$$

$$
\frac{m_W}{m_e}
=
\frac{y_e^-1}{R}
\left(1+\frac{1}{\Psi_v}\right),
\qquad
m_Z=\frac{m_W}{\cos\theta_W},
$$

$$
\alpha_s(m_Z)=1-\frac{m_W}{m_Z}.
$$

````python
# Higgs-sector base structure and fixed structural integers.
B_H   = 12 * R * alpha2_transfer(alpha, 1, 2)
Psi_H = 12 * (4 * (3 * R**2) + 3)
Psi_v = 3**2 * Psi_H

# Higgs and vacuum scales.
mH_over_me = (1 / 2.0) * B_H + Psi_H
v_over_me  = B_H - Psi_v
mH_GeV     = mH_over_me * me_gev
v_GeV      = v_over_me  * me_gev

# Electron Yukawa and weak-sector values.
ye = math.sqrt(2) / v_over_me
sin2_thetaW = 2 * math.pi * alpha * R**2 * (1 + 3**-3)
cos_thetaW  = math.sqrt(1 - sin2_thetaW)
mW_over_me  = (1 / ye) / R * (1 + 1 / Psi_v)
mW_GeV      = mW_over_me * me_gev
mZ_GeV      = mW_GeV / cos_thetaW
alpha_s_mZ  = 1 - mW_GeV / mZ_GeV

print_table([
    ("B_H", B_H, "", ".12f"),
    ("Psi_H", Psi_H, "", ".0f"),
    ("Psi_v", Psi_v, "", ".0f"),
    ("m_H / m_e", mH_over_me, "", ".12f"),
    ("v / m_e", v_over_me, "", ".12f"),
    ("m_H", mH_GeV, "GeV", ".12f"),
    ("v", v_GeV, "GeV", ".12f"),
    ("y_e", ye, "", ".12e"),
    ("sin^2(theta_W)", sin2_thetaW, "", ".12f"),
    ("m_W / m_e", mW_over_me, "", ".12f"),
    ("m_W", mW_GeV, "GeV", ".12f"),
    ("m_Z", mZ_GeV, "GeV", ".12f"),
    ("alpha_s(m_Z)", alpha_s_mZ, "", ".12f"),
])
````
